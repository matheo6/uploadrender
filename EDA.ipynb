### Importacion de librerias y lectura de archivos csv 

```python
import pandas as pd 
credits= pd.read_csv("credits.csv")
movies_Dataset=pd.read_csv("movies_dataset.csv",low_memory=False)
credits['index']=credits['id']
credits.set_index('id',inplace=True)
movies_Dataset['index']=movies_Dataset['id']
movies_Dataset.set_index('id',inplace=True,)
```

### Definicion de Funcion para desanidar las columnas contraidas en listas y diccionarios 

```python
def subtract(df,colum):
    df[colum]=df[colum].apply(lambda x: eval(x))
    tempdf=df[colum].explode()
    finaldf=pd.json_normalize(tempdf)
    finaldf['index']=tempdf.index
    return finaldf
```

### Generacion de nuevos dataframes de las columnas anidadas llamando a la funcion subtrac

```python
crew=subtract(cleanCredits,'crew')
cast=subtract(cleanCredits,'cast')
genres=subtract(moviesclean,'genres')
```
